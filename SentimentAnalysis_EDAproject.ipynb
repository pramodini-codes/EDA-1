```python
import pandas as pd
import nltk
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import string
from nltk.tokenize import word_tokenize
```

```python
df=pd.read_excel("/content/sentiment analysis.xlsx")
df
```

```python
stop_words=stopwords.words('english')
ps=PorterStemmer()
```

```python
def preprocessing(text):
  text=text.lower()
  token=nltk.word_tokenize(text)
  new=[]
  for i in token:
    if i not in stop_words and i not in string.punctuation:
      stemmer=ps.stem(i)
      new.append(stemmer)
  return " ".join(new)
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
df["clean"]=df["text"].apply(preprocessing)
```

```python
df["clean"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vector=TfidfVectorizer()
X=vector.fit_transform(df["clean"])
y=df["label"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=55)
```

```python
log=LogisticRegression()
log.fit(X_train,y_train)
```

```python
print("train accuracy",log.score(X_train,y_train))
print("test accuracy",log.score(X_test,y_test))
```

```python

def pred_text(text):
  text=text.lower()
  text=preprocessing(text)
  vec=vector.transform([text])
  predict=log.predict(vec)[0]
  if predict==1:
    return "positive"
  else:
    return "negative"
```

```python
print(pred_text("i hate this enjoyment"))
```

```python
print(pred_text("i love this enjoyment"))
```

```python
print(pred_text("I'm feeling really down today."))
```
