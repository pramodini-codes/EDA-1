```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('/content/Diwali Sales Data.csv',encoding='unicode_escape')
```

After uploading the file, you will need to re-run the cell that reads the CSV file to load the data into the DataFrame.

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Marriage'})
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sale_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sale_gen)
```

```python
ax=sns.countplot(x='Age',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sale_gen=df.groupby(['Age'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age',y='Amount',data=sale_gen)
```

```python
sale_gen=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sale_gen)
```

```python
ax=sns.countplot(x='State',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

```python
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sale_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sale_state,hue='Gender')
```

```python
ax=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(30,20)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sale_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=sale_occ)
```

```python
ax=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(20,3)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sale_occ=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(28,5)})
sns.barplot(x='Product_Category',y='Amount',data=sale_occ)
```

```python
sale_occ=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(28,5)})
sns.barplot(x='Product_ID',y='Orders',data=sale_occ)
```
